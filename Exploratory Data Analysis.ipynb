## Exploratory Data Analysis

We will now explore the dataset that we cleaned up. We want to find out the relationships between students' final grades (G3) and the rest of the other attributes that we picked out to get an idea of how these variables can affect G3. This will become helpful when we build our prediction models.

### Importing essential libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import plotly.graph_objects as go
import statsmodels.api as sm
from pyecharts import options as opts
from pyecharts.charts import Bar

import warnings
warnings.filterwarnings('ignore')
mpl.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False 
```

```
student_data = pd.read_csv('student_data.csv')
```

### Data Analysis

```
student_data.head()
```

```
student_data.school.value_counts()
```

```
fig = plt.figure(figsize=(10, 8))
sns.set_style("dark",{"font.sans-serif":["Arial"]})
sns.boxplot(data=student_data, x='school', y='G3')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('School Name', fontsize = 20)
plt.ylabel('Grade', fontsize = 20)
plt.title('Box plot of grades of students from different schools', fontsize = 20)
plt.show()
```

```
fig = plt.figure(figsize=(10, 8))
sns.set_style("dark",{"font.sans-serif":["Arial"]})
sns.boxplot(data=student_data, x='sex', y='G3')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Sex', fontsize = 20)
plt.ylabel('Grade', fontsize = 20)
plt.title('Box plot of grades of students of different sexes', fontsize = 20)
plt.show()
```

```
fig = plt.figure(figsize=(10, 8))
sns.set_style("dark",{"font.sans-serif":["Arial"]})
sns.violinplot(data=student_data, x='age', y='G3')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Age', fontsize = 20)
plt.ylabel('Grade', fontsize = 20)
plt.title('Violin Plot of grades of students of different ages', fontsize = 20)
plt.grid()
plt.show()
```

```
student_data.G3.value_counts()
```

```
# Classify students' grades into four categories: A, B, C, D
student_data['level'] = pd.cut(student_data.G3, bins=[0,5,10,15,20], labels=[ 'D', 'C', 'B', 'A'])

student_data.head()
```

```
student_data.level.value_counts()
```

```
# Creating pie chart
fig = go.Figure(data=[go.Pie(labels=student_data.level.value_counts().index, values=student_data.level.value_counts().values, hole=.3)])

# Setting title of pie chart
fig.update_layout(title="Pie Chart of Categories of Students' Grades")

# Show figure
fig.show()
```

```
fig = plt.figure(figsize=(10, 8))
sns.set_style("dark",{"font.sans-serif":["Arial"]})
sns.boxplot(data=student_data, x='traveltime', y='G3')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Average Travel Time/h', fontsize = 20)
plt.ylabel('Grade', fontsize = 20)
plt.title('Box Plot of grades of students of different average travel time/h', fontsize = 20)
plt.grid()
plt.show()
```

```
fig = plt.figure(figsize=(10, 8))
sns.set_style("dark",{"font.sans-serif":["Arial"]})
sns.boxplot(data=student_data, x='studytime', y='G3')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Average Time Spent in Studying/h', fontsize = 20)
plt.ylabel('Grade', fontsize = 20)
plt.title('Box Plot of Students Grade of Different Average Time Spent in Studying', fontsize = 20)
plt.grid()
plt.show()
```

```
famrel_index_list = [1, 2, 3, 4, 5]
famrel_mean_sorce = []
for i in famrel_index_list:
    famrel_mean_sorce.append(student_data[student_data.famrel==i].G3.mean())
# Creating bar chart
fig = go.Figure(data=[go.Bar(x=famrel_index_list, y=famrel_mean_sorce)])

# Setting layout of bar chart
fig.update_layout(title='Bar Chart of Students Average Grade with Different Family Relationships', xaxis_title='Family Relationships', yaxis_title='Average Grade')

# Show Figure
fig.show()
```

```
freetime_index_list = [1, 2, 3, 4, 5]
freetime_mean_sorce = []
for i in freetime_index_list:
    freetime_mean_sorce.append(student_data[student_data.freetime==i].G3.mean())

fig = go.Figure(data=[go.Bar(x=freetime_index_list, y=freetime_mean_sorce)])

fig.update_layout(title='Bar Chart of Students Average Grade with Different Amount of Free Time', xaxis_title='Free Time', yaxis_title='Avarage Grade')

fig.show()
```

```
goout_index_list = [1, 2, 3, 4, 5]
goout_mean_sorce = []
for i in goout_index_list:
    goout_mean_sorce.append(student_data[student_data.goout==i].G3.mean())
bar = Bar()
bar.add_xaxis(goout_index_list) 
bar.add_yaxis("Grade",[float(i) for i in goout_mean_sorce]) 
bar.set_global_opts(title_opts=opts.TitleOpts(title='Bar Chart of Students with Different Average Time Spent in Going Out',subtitle=''), 
xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-15)),)  
bar.render_notebook()
```

---

##### From the above graph we can conclude that when Semester GPA is greater than approximately 4.0, the correlation betwee Final Grade and Semester GPA is strongly positive.

##### As a result, based on the exploratory data analysis, we found out that age, Medu, Fedu, traveltime, studytime, Dalc, Walc, health, absences, G1, G2 are the factors that have greater effects on Students' Grades (G3).
